# Demonstrate multiple reduce operators in groupby aggregation

```python
from collections import OrderedDict

from pandas.util.testing import assert_frame_equal

np.random.seed(1)

df = OrderedDict([('keys', np.random.randint(1, 10, size=20).astype('uint32')),
                  ('values', (np.random.randint(1, 10, size=20) * .71).astype('float32')),
                  ('values2', np.random.randint(1, 10, size=20).astype('uint32'))])
ddf = DeviceDataFrame(df, context=context)
g = ddf.groupby(['keys'])
reduce_op = ['sum', 'max']
print 'Input'
display(ddf.df.T)
print 'Returned new data frame'
reduced_out = g.agg(reduce_op=reduce_op)
display(reduced_out.df.T)

# Use existing output frame.
print 'Use existing output frame.'
reduced_out.reset_views()
ddf_out = DeviceDataFrame(reduced_out.df * 0)
g.agg(reduce_op=reduce_op, out=ddf_out).df.T

a, b = ddf_out.df.T, g.ref_agg(reduce_op).T
assert_frame_equal(a, b)
```
