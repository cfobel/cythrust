```python
%load_ext cythonmagic
import pkg_resources

cythrust_root = pkg_resources.resource_filename('cythrust', '')
```

```python
from IPython.display import display
import numpy as np
import pandas as pd
from collections import OrderedDict
import cythrust.device_vector as dv
from cythrust import DeviceDataFrame
```

```python
%%cython -f --cplus -I$cythrust_root --compile-args=-DTHRUST_DEVICE_SYSTEM=THRUST_DEVICE_SYSTEM_CPP

from libc.stdint cimport int32_t, bool
from cythrust.thrust.partition cimport (counted_partition,
                                        counted_partition_w_stencil)
from cythrust.thrust.functional cimport positive, non_negative
from cythrust.thrust.tuple cimport (make_tuple2, make_tuple3,
                                    make_tuple4, make_tuple5)
from cythrust.thrust.iterator.zip_iterator cimport make_zip_iterator
from cythrust.device_vector cimport DeviceVectorViewInt32


def pack_positive(DeviceVectorViewInt32 values1,
                  DeviceVectorViewInt32 values2,
                  DeviceVectorViewInt32 stencil,
                  bint pack_stencil=True):
    cdef non_negative[int32_t] filter_f
    cdef int32_t count =  counted_partition_w_stencil(
            make_zip_iterator(make_tuple2(values1._begin,
                                          values2._begin)),
            make_zip_iterator(make_tuple2(values1._end,
                                          values2._end)),
            stencil._begin, filter_f)
    if pack_stencil:
        counted_partition(stencil._begin, stencil._end, filter_f)
    return count
```

```python
frame = DeviceDataFrame(OrderedDict([
    ('values1', np.arange(10, dtype=np.int32)),
    ('values2', np.arange(10, 0, -1, dtype=np.int32)),
    ('stencil', np.random.randint(2, size=10)
     .astype(dtype=np.int32) - 1)]))
frame[:].T
```

```python
pack_positive(frame.v['values1'], frame.v['values1'],
              frame.v['stencil'])
f2 = frame.view(3, 5)
frame[:].T
```
